```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Olympics.csv')
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['medal'].fillna(value=0, inplace=True)
```

```python
df['height'].fillna(method='bfill', inplace=True)
```

```python
df['weight'].fillna(method='bfill', inplace=True)
```

```python
df.isnull().sum()
```

```python
athlete_medals = df[['name', 'medal']]
athlete_medals
```

```python
athlete_counts = df['name'].value_counts()
non_duplicated_athletes = athlete_counts[athlete_counts == 1].index.tolist()
num_non_duplicated_athletes = len(non_duplicated_athletes)
print("The number of non duplalicsted athletes: ",num_non_duplicated_athletes)

```

```python
gold_medal_athletes = df[df['medal'] == 'Gold']
number_of_athletets_win_gold_medals=len(gold_medal_athletes)
print("The number of athletets_win_gold_medals: ",number_of_athletets_win_gold_medals)
```

```python
silver_medal_athletes = df[df['medal'] == 'Silver']
silver_ordered_by_year = silver_medal_athletes.sort_values(by='year')
silver_ordered_by_year
```

```python
medal_counts = df.groupby('team')['medal'].value_counts().unstack().fillna(0)
medal_counts
```

```python
most_medals_athlete = df['name'].value_counts().idxmax()
print("The most medakl wiun by the athlete: ",most_medals_athlete)
```

```python
freestyle_events = df[df['event'].str.contains('Freestyle', case=False)]['event'].unique()
print("The Freestyle Event is : ",freestyle_events)
```

```python
top_3_athletes_by_sport = df.groupby(['sport', 'name'])['medal'].count().reset_index(name='medal_count')
top_3_athletes_by_sport = top_3_athletes_by_sport.sort_values(['sport', 'medal_count'], ascending=[True, False]).groupby('sport').head(3)
top_3_athletes_by_sport
```

```python
athletes_multiple_medals = df.groupby(['year', 'name'])['medal'].count()
athletes_multiple_medals = athletes_multiple_medals[athletes_multiple_medals > 1].reset_index(name='medal_count')
athletes_multiple_medals
```

```python
df.head()
```

```python
summer_gold = df.loc[(df['season'] == 'Summer') & (df['medal'] == 'Gold'), 'team'].unique()
winter_gold = df.loc[(df['season'] == 'Winter') & (df['medal'] == 'Gold'), 'team'].unique()
countries_winning_both = set(summer_gold) & set(winter_gold)
countries_winning_both
```

```python
first_medal_year = df.groupby('team')['year'].min()
last_medal_year = df.groupby('team')['year'].max()
year_difference = last_medal_year - first_medal_year
year_difference
```

```python
silver_athletes = df[df['medal'] == 'Silver'][['name', 'event', 'year']]
gold_athletes = df[df['medal'] == 'Gold'][['name', 'event', 'year']]

silver_gold_athletes = silver_athletes.merge(gold_athletes, on=['name', 'event'], suffixes=('_silver', '_gold'))
silver_gold_athletes
```

```python
countries_medals = df.groupby('noc')['sport'].nunique()
countries_with_more_than_8_sports = countries_medals[countries_medals > 8]
countries_with_more_than_8_sports
```

```python
countries_gold_medals = df[df['medal'] == 'Gold'].groupby('noc').size()
countries_with_more_than_4_gold = countries_gold_medals[countries_gold_medals > 4]
countries_with_more_than_4_gold
```
